```python
df1=pd.read_csv('/content/data_sales.csv')
df1.head()
```

```python
# Append a new variable =.
df1['Amount'] = df1['Unit Price']*df1['Units']
df1.head()
```

```python
#  Average unit price for each region (using groupby)
df1.groupby('Region')['Unit Price'].mean()
```

```python
# Average unit price for each region (using pivot table)
df1.pivot_table(index = 'Region', values = 'Unit Price', aggfunc=np.mean)
```

```python
# Average unit price and units for each region in one code sentence using groupby
df1.groupby('Region')[['Unit Price','Units']].mean()
```

```python
# Average unit price and units for each region in one code sentence using pivot_table

df1.pivot_table(index='Region', values = ['Unit Price','Units'], aggfunc=np.mean)
```

```python
  # Total units for each region and item type in one code sentence. Fill missing values with 0
  df1.pivot_table(index=['Region', 'Item'], values = 'Units', aggfunc=np.sum, fill_value=0 )
```

```python
 # Total sales amount for each region and item type in one code sentence. Use the pivot_table() method. Fill the missing values with 0
 df1.pivot_table(index=['Region', 'Item'], values ='Amount', aggfunc=np.sum, fill_value=0 )
```

```python
df2=pd.read_csv('/content/data_coffeeshop.csv')
df2.head()
```

```python
 # Make a frequency table of 'yearOfStart' and visualize by year.
#  Sort by the year.
import matplotlib.pyplot as plt
# Draw a line plot from 1997 to 2014.
table =df2['Year of Start'].value_counts().sort_index()
table
plt.plot(table.index, table.values)
plt.xlim([1997,2014])
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your DataFrame is named df2

# Group data by 'Current State' and 'Year of Start', then count frequencies
in_business = df2[df2['Current State'] == 'In'].groupby('Year of Start')['Year of Start'].count().sort_index()
out_business = df2[df2['Current State'] == 'Out'].groupby('Year of Start')['Year of Start'].count().sort_index()

# Create the plot
plt.plot(in_business.index, in_business.values, label='In Business')
plt.plot(out_business.index, out_business.values, label='Out of Business')

# Set plot limits and labels
plt.xlim([1997, 2014])
plt.xlabel('Year of Start')
plt.ylabel('Frequency')
plt.title('Yearly Trend of Coffee Shop Starts')
plt.legend()

# Show the plot
plt.show()
```
